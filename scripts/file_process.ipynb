```python
# delete half images
import os
import shutil

# Define the directory containing the images
img_dir = "/home/doan/VA_0223/eval/data/processed/tanks_and_temples/Courthouse/images"

# Get all image files
image_files = [f for f in os.listdir(img_dir) 
             if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

# Delete half of the images
for i in range(len(image_files)):
    if int(image_files[i].split(".")[0]) % 2 == 0:
        print(int(image_files[i].split(".")[0]))
        os.remove(os.path.join(img_dir, image_files[i]))

print(f"Deleted {len(image_files) // 2} images")
```

```python
images_folder = '/home/doan/VA_0223/eval/data/processed/tanks_and_temples/Barn/images'
images_test_folder = '/home/doan/VA_0223/eval/data/processed/tanks_and_temples/Barn/images_test'
test_file = '/home/doan/VA_0223/eval/data/processed/tanks_and_temples/Barn/test.txt'

# copy images to images_test_folder
os.makedirs(images_test_folder, exist_ok=True)
with open(test_file, 'r') as f:
    for line in f:
        image_name = line.strip()
        src_path = os.path.join(images_folder, image_name)
        dst_path = os.path.join(images_test_folder, image_name)
        shutil.copy(src_path, dst_path)

```

```python
## copy test images for all scenes
root_dir = '/home/doan/VA_0223/eval/data/processed/tanks_and_temples'
for scene in os.listdir(root_dir):
    if os.path.isdir(os.path.join(root_dir, scene)):
        images_folder = os.path.join(root_dir, scene, 'images')
        images_test_folder = os.path.join(root_dir, scene, 'images_test')
        test_file = os.path.join(root_dir, scene, 'test.txt')
        
        # copy images to images_test_folder
        os.makedirs(images_test_folder, exist_ok=True)
        with open(test_file, 'r') as f:
            for line in f:
                image_name = line.strip()
                src_path = os.path.join(images_folder, image_name)
                dst_path = os.path.join(images_test_folder, image_name)
                shutil.copy(src_path, dst_path)

```
